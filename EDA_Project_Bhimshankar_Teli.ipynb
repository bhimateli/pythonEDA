# Treding Line chart for Top 10 countries

```python
top10_countries = df.groupby('Country')['Total Water Consumption (Billion m3)'] \
                    .mean() \
                    .sort_values(ascending=False) \
                    .head(10) \
                    .index

df_top10 = df[df['Country'].isin(top10_countries)]

plt.figure(figsize=(10,6))

for country in top10_countries:
    country_data = df_top10[df_top10['Country'] == country]
    country_data = country_data.sort_values('Year')
    
    plt.plot(country_data['Year'],
             country_data['Total Water Consumption (Billion m3)'],
             marker='o',
             label=country)

plt.title("Top 10 Countries Water Consumption Trend")
plt.xlabel("Year")
plt.ylabel("Water Consumption (Billion m3)")
plt.legend()

plt.figtext(
    0.5, -0.12,
    "Interpretation: The line chart shows the water consumption trends for the top 10 countries "
    "over the years. China consistently records the highest water consumption, followed by India "
    "and the United States. Most countries display a steady upward trend, indicating increasing "
    "water demand over time. Even countries with lower consumption levels such as Angola and "
    "Slovakia show gradual growth, highlighting an overall rise in global water usage.",
    wrap=True,
    horizontalalignment='center',
    fontsize=10
)

plt.show()
```

```python
top_countries = df.groupby('Country')['Total Water Consumption (Billion m3)'] \
                  .mean() \
                  .sort_values(ascending=False) \
                  .head(2) \
                  .index

df_top = df[df['Country'].isin(top_countries)]

for country in top_countries:
    
    country_data = df_top[df_top['Country'] == country]
    
    plt.figure(figsize=(10,6))
    
    plt.plot(country_data['Year'], country_data['Agricultural Water Use (%)'], label="Agriculture")
    plt.plot(country_data['Year'], country_data['Industrial Water Use (%)'], label="Industrial")
    plt.plot(country_data['Year'], country_data['Household Water Use (%)'], label="Household")
    
    plt.title(f"Water Usage by Sector Over Time - {country}")
    plt.xlabel("Year")
    plt.ylabel("Water Usage (%)")
    plt.legend()
    plt.figtext(
        0.5, -0.12,
        f"Interpretation: The chart illustrates the sector-wise water usage trend in {country} over time. "
        "Agriculture accounts for the largest share of water consumption throughout the years, although "
        "its proportion shows a slight decline. In contrast, industrial and household water usage display "
        "a gradual increase, reflecting growing industrial activities and rising domestic demand. "
        "Overall, the trend indicates a gradual shift in water usage distribution while agriculture "
        "continues to dominate total water consumption.",
        wrap=True,
        horizontalalignment='center',
        fontsize=10
    )
    
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(10,5))

sns.boxplot(
    data=df,
    x='Water Scarcity Level',
    y='Total Water Consumption (Billion m3)'
)

# Calculate statistics per category
stats = df.groupby('Water Scarcity Level')['Total Water Consumption (Billion m3)'].agg(
    Min='min',
    Q1=lambda x: x.quantile(0.25),
    Median='median',
    Q3=lambda x: x.quantile(0.75),
    Max='max'
)

stats['IQR'] = stats['Q3'] - stats['Q1']

# Draw lines for each category
for i, level in enumerate(stats.index):

    median = stats.loc[level, 'Median']
    q1 = stats.loc[level, 'Q1']
    q3 = stats.loc[level, 'Q3']
    min_val = stats.loc[level, 'Min']
    max_val = stats.loc[level, 'Max']

    plt.scatter(i, median, color='green', s=80)
    plt.text(i, median, f'Median:{median:.1f}', ha='center', va='bottom')

    plt.scatter(i, q1, color='blue')
    plt.scatter(i, q3, color='blue')

    plt.scatter(i, min_val, color='purple')
    plt.scatter(i, max_val, color='purple')

plt.title("Consumption vs Scarcity Level")
plt.xlabel("Water Scarcity Level")
plt.ylabel("Total Water Consumption (Billion m3)")

plt.figtext(
    0.5, -0.15,
    "Interpretation: The box plot compares total water consumption across different water scarcity levels. "
    "Most observations show relatively lower consumption values within each category, while numerous outliers "
    "indicate countries with significantly higher water usage. The 'High' and 'Moderate' scarcity categories "
    "display wider variability and several extreme consumption values, suggesting that regions experiencing "
    "higher scarcity may still have large water demands. In contrast, the 'Critical' category shows a smaller "
    "range of consumption values, indicating relatively lower variability in water usage.",
    wrap=True,
    horizontalalignment='center',
    fontsize=10
)

plt.tight_layout()
plt.show()

print(stats)
```

```python
#Three sector percentages. list comprehension
df['Total_Sector_Use'] = [  
    a + b + c
    for a, b, c in zip(
        df['Agricultural Water Use (%)'],
        df['Industrial Water Use (%)'],
        df['Household Water Use (%)']
    )
]
df['Total_Sector_Use'].head(5)
```

```python
df['Consumption Category'] = df[
    'Total Water Consumption (Billion m3)'
].apply(lambda x: 'High' if x > 600 else 'Low')
```

```python
df['Consumption Level'] = df[
    'Total Water Consumption (Billion m3)'
].apply(
    lambda x:
        'Low' if x < 400
        else 'Medium' if x < 700
        else 'High'
)
```

# User-Defined Functions:
 Write custom functions to perform repetitive tasks or complex
calculations.
 Example: Create a function to calculate the range of salary as low,
medium, high

```python
def consumption_category(value):
    if value < 400:
        return "Low"
    elif value < 700:
        return "Medium"
    else:
        return "High"

consumption_category(1000)
```

```python
def water_stress(row):
    if row['Groundwater Depletion Rate (%)'] > 60:
        return "Severe"
    elif row['Groundwater Depletion Rate (%)'] > 40:
        return "Moderate"
    else:
        return "Low"
df['Stress Category'] = df.apply(water_stress, axis=1)
```

```python
def categorize_levels(value, low, high):
    if value < low:
        return "Low"
    elif value < high:
        return "Medium"
    else:
        return "High"
df['Per Capita Category'] = df[
    'Per Capita Water Use (L/Day)'
].apply(lambda x: categorize_levels(x, 150, 300))

```
